- Determine the probability that a package will arrive in 1 day if it traveled less than 1000 miles. [1pts]

- Determine the probability that if a package arrived in 3 days, it traveled greater than 1000 miles but less than 2000 miles.. [1pts]


### Answer
### 5.
#### 1.
$P\{miles \in (2000, 3000)\} = 1-P\{miles \in (0,2000)\} = 1-85\% = 15\%$

#### 2.
$P\{1\,day\} = P\{1\,day, (0, 1000)\} + P\{1\,day, (1000, 2000)\} + P\{1\,day, (2000, 3000)\} = 10\% $\
$P\{1\,day, (0, 1000)\} =  P\{1\,day, (1000, 2000)\}$, 
$P\{1\,day, (2000, 3000)\} = 0 $\
$\rightarrow P\{1\,day, (0, 1000)\} = P\{1\,day, (1000, 2000)\} = 5\%$\
\
$P\{(1000, 2000)\} = P\{(0, 2000)\} - P\{(0, 1000)\} = 85\% - 50\% = 35\%$\
\
$P\{1\,day|(1000, 2000)\} = \dfrac{P\{1\,day, (1000, 2000)\}}{P\{(1000, 2000)\}} = \dfrac{5\%}{35\%} = \dfrac{1}{7}$

#### 3.
$P\{1\,day, (0, 1000)\} = 5\%$\
$P\{(0, 1000)\} = 50\%$\
$P\{1\,day|(0, 1000)\} = \dfrac{P\{1\,day, (0, 1000)\}}{P\{(0, 1000)\}} = \dfrac{5\%}{50\%} = \dfrac{1}{10}$

#### 4.
$P\{(1000, 2000)|3\,day\} = \dfrac{P\{3\,day, (1000,2000)\}}{P\{3\,day\}} = \dfrac{P\{3\,day|(1000,2000)\}P\{(1000,2000)\}}{P\{3\,day\}} = \dfrac{10\%\times35\%}{1-10\%-65\%}=14\%$
